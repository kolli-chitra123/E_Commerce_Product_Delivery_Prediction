***TRAIN TEST SPLIT***

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Reached.on.Time_Y.N', axis=1), df['Reached.on.Time_Y.N'], test_size=0.2, random_state=0)
```

**DECISION TREE**

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'max_depth': [2,4,6,8],
    'min_samples_leaf': [2,4,6,8],
    'min_samples_split': [2,4,6,8],
    'criterion': ['gini', 'entropy'],
    'random_state': [0,42]}
grid = GridSearchCV(estimator=dtc, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='accuracy')
grid.fit(X_train, y_train)
print('Best parameters: ', grid.best_params_)
```

```python
dtc = DecisionTreeClassifier(criterion='gini', max_depth=6, min_samples_leaf=6, min_samples_split=2, random_state=0, class_weight='balanced')
dtc.fit(X_train, y_train)
```

```python
print('Training accuracy: ', dtc.score(X_train, y_train))
```

```python
dtc_pred = dtc.predict(X_test)
```

**LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='liblinear',max_iter=500)
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_train,y_train)
```

```python
lr_pred=lr.predict(X_test)
```

**K NEAREST NEIGHBOR**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
knn_pred = knn.predict(X_test)
```

**EVALUATION**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, mean_absolute_error, r2_score, mean_squared_error
```

```python
ig, ax = plt.subplots(1,3,figsize=(15,10))
sns.heatmap(confusion_matrix(y_test, dtc_pred), annot=True, cmap='coolwarm', ax=ax[0]).set_title('Decision Tree Classifier')
sns.heatmap(confusion_matrix(y_test, lr_pred), annot=True, cmap='coolwarm', ax=ax[1]).set_title('Logistic Regression')
sns.heatmap(confusion_matrix(y_test, knn_pred), annot=True, cmap='coolwarm', ax=ax[2]).set_title('KNN Classifier')
```

**CLASSIFICATION REPORT**

```python
print('Decision Tree Classifier: \n', classification_report(y_test, dtc_pred))
print('Logistic Regression: \n', classification_report(y_test, lr_pred))
print('KNN Classifier: \n', classification_report(y_test, knn_pred))
```

**Accuracy Percentage**

```python
print('Decision Tree Classifier: \n', int(accuracy_score(y_test, dtc_pred)*100))
print('Logistic Regression: \n', int(accuracy_score(y_test, lr_pred)*100))
print('KNN Classifier: \n', int(accuracy_score(y_test, knn_pred)*100))
```

**MODEL COMPARISION**

```python
models = ['Decision Tree Classifier', 'Logistic Regression', 'KNN Classifier']
accuracy = [accuracy_score(y_test, dtc_pred), accuracy_score(y_test, lr_pred), accuracy_score(y_test, knn_pred)]
sns.barplot(x=models, y=accuracy, hue=models, palette='magma', dodge=False, legend=False).set_title('Model Comparison')
plt.xticks(rotation=90)
plt.ylabel('Accuracy')
```
